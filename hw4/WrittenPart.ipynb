```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt

# clf = LinearSVC(random_state=0)
clf = SVC(kernel='linear', C=10)
training_data = [[3,4],[2,2],[4,4],[1,4],[2,1],[4,3],[4,1]]
training_label = [1,1,1,1,0,0,0]
clf.fit(training_data,training_label)
b0=clf.intercept_[0]
b1,b2=clf.coef_[0]
margin = 1/(b1**2+b2**2)**0.5
print(b0, b1, b2)
print(margin)
```

```python
import numpy as np

redDots=[[3, 4], [2, 2], [4, 4], [1, 4]]
blueDots=[[2, 1], [4, 3], [4, 1]]

plt.plot( [red[0] for red in redDots],[red[1] for red in redDots],'ro',color='red')
plt.plot( [blue[0] for blue in blueDots],[blue[1] for blue in blueDots],'ro',color='blue')
x1 = np.linspace(0,5)
x2 = (-b0 - b1 * x1) / b2 
lb = x2 - np.sqrt(1 + (b1/b2) ** 2) * margin
ub = x2 + np.sqrt(1 + (b1/b2) ** 2) * margin
plt.plot(x1,x2, color = 'black')
plt.plot(x1,ub, color = 'orange', linestyle='--')
plt.plot(x1,lb, color = 'orange', linestyle='--')
plt.show()
```

```python
def relu(x):
    return np.maximum(x,0)    

# Define weights and biases
W1 = np.array([2,5/3,5/3,11/3,5/3]) # [2,5/3,5/3,11/3,5/3]
b1 = np.array([-2,-10/3,-25/3,-22,-15]) # [-2,-10/3,-25/3,-22,-15]
W2 = np.array([1,-1,1,-1,1])
b2 = 0

# Input array
x = np.linspace(0, 10, 100)

# Forward propagation
resultArr = []
for i in range(len(x)):
    z1 = np.array(x[i]).dot(W1) + b1
    a1 = relu(z1)
    a2 = a1.dot(W2) +b2
    resultArr.append(relu(a2))
    

# Generate graph and plot results

plt.plot(x,resultArr)
plt.axis([0, 10, -1, 6])
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('"Pride Rock"')
plt.grid(True)
plt.show()
```

```python
W1 = np.array([2,5/3,5/3,11/3,5/3]) # slope change absolute values [2,5/3,5/3,11/3,5/3]
b1 = -np.array([1,2,5,6,9])*W1 # [-2,-10/3,-25/3,-22,-15]
W2 = np.array([1,-1,1,-1,1]) # slope change direction
b2 = 0
x = np.linspace(0, 10, 100)

mapArr = []
for i in range(len(x)):
    z1 = np.array(x[i]).dot(W1) + b1
    a1 = relu(z1)
    a2 = a1.dot(W2) +b2
    mapArr.append(a2)
    
plt.plot(x,mapArr)
plt.axis([0, 10, -1, 6])
plt.xticks(np.arange(0, 10, 1))
plt.grid(linewidth=0.5)
plt.title('"Pride Rock"')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```
